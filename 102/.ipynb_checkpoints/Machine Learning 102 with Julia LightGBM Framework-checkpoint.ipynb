## Input missing values

```julia
#Make function to input missing value in age column, this function ensure that distribution of age column will not change
#even based on survival criteria
surv_age = combine(groupby(surv, :Age), :Age=>length)
surv_age_clean = filter!(row->!ismissing(row.Age), surv_age)
begin
    surv_age_clean[!,:proportion] = surv_age_clean[:Age_length]/sum(Array(surv_age_clean.Age_length))
end
not_surv_age = combine(groupby(not_surv, :Age), :Age=>length)
not_surv_age_clean = filter!(row->!ismissing(row.Age), not_surv_age)
begin
    not_surv_age_clean[!,:proportion] = not_surv_age_clean[:Age_length]/sum(Array(not_surv_age_clean.Age_length))
    end;
```

```julia
#Input the missing value in age column
age_surv = Array(surv_age_clean.Age)
weights_surv = Array(surv_age_clean.proportion)
age_fill = sample(age_surv, Weights(weights_surv), sum((ismissing.(data_clean[:Age])).&(isequal.(data_clean[:Survived],1))))
begin
    data_clean[(ismissing.(data_clean[:Age])).&(isequal.(data_clean[:Survived],1)),:Age] = age_fill
end

age_not_surv = Array(not_surv_age_clean.Age)
weights_not_surv = Array(not_surv_age_clean.proportion)
age_fill = sample(age_not_surv, Weights(weights_not_surv), sum((ismissing.(data_clean[:Age])).&(isequal.(data_clean[:Survived],0))))
begin
    data_clean[(ismissing.(data_clean[:Age])).&(isequal.(data_clean[:Survived],0)),:Age] = age_fill
    end;
```

```julia
#check the data
column_name = names(data_clean)
nmissing_values =  Array([])
percentage_missing = Array([])
for i in column_name
    nMissings = length(findall(x -> ismissing(x), data_clean[!,i]))
    append!(nmissing_values, nMissings)
    append!(percentage_missing, nMissings/nrow(data))
end


missing_data =  DataFrame(columns_name=column_name, num_missing=nmissing_values, percentage=percentage_missing)
```

# Modeling

```julia
using ScikitLearn
using ScikitLearn.CrossValidation: train_test_split #using scikitleran to split the data
@sk_import metrics: classification_report;
```

```julia
#change categorical variable into numerical
#this is manual label encoder, we not use hot encoder because lgbm is tree based model
data_clean.Sex = replace.(data_clean.Sex, Ref("male" => 1))
data_clean.Sex = replace.(data_clean.Sex, Ref("fe1" => 0))
data_clean.Embarked = replace.(data_clean.Embarked, Ref("S" => 0))
data_clean.Embarked = replace.(data_clean.Embarked, Ref("C" => 1))
data_clean.Embarked = replace.(data_clean.Embarked, Ref("Q" => 2))
first(data_clean, 5)
```

```julia
#change categorical variable into numerical
data_clean[:Sex] = parse.(Int,data_clean[:Sex])
data_clean[:Embarked] = parse.(Int,data_clean[:Embarked])
first(data_clean, 5)
```

```julia
X = convert(Array,data_clean[:,2:end]) #Change data into array
y = convert(Array,data_clean[:,1]);
```

```julia
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2,
stratify=y, random_state=42); #split the data
```

```julia
using LightGBM
using LightGBM: fit!
using LightGBM: predict
```

```julia
estimator = LGBMClassification(
    objective = "binary",
    num_iterations = 1000,
    learning_rate = .1,
    early_stopping_round = 5,
    feature_fraction = .8,
    bagging_fraction = .9,
    bagging_freq = 1,
    num_leaves = 1000,
    num_class = 1,
    metric = ["auc", "binary_logloss"]
) #arbitary parameter, u can tuning this parameter
```

```julia
fit!(estimator, X_train, y_train, (X_test, y_test)) #fit the data
```

```julia
y_pred = predict_classes(estimator, X_test; binary_threshold = 0.7)
println(classification_report(y_test, y_pred))
```
