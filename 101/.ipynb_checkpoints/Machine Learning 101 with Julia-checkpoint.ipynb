```julia
#Read data
data = DataFrame(CSV.File("iris.data")) #Same as pd.read_csv
first(data, 5) #Same as data.head(5)
```

```julia
#Get summary statistics from all feature
describe(data[:,[1,2,3,4]])
```

```julia
#Create histogram plot to know distribution each feature
p1 = histogram(data.sepal_length, title="Sepal Length")
p2 = histogram(data.sepal_width, title="Sepal Width")
p3 = histogram(data.petal_length, title="Petal Length")
p4 = histogram(data.petal_width, title="Petal Width")
plot(p1,p2,p3,p4, layout=(2,2), legend=false)
plot!(size=(950,400))
```

```julia
#Plot the scatter to know the data well seperated or not
s1 = scatter(data.sepal_length, data.sepal_width, title="Scatter Sepal Length and Width",
xlabel = "Sepal Length", ylabel = "Sepal Width", group=data.class)
s2 = scatter(data.petal_length, data.petal_width, title="Scatter Petal Length and Width",
xlabel = "Petal Length", ylabel = "Petal Width", group=data.class)
plot(s1,s2, layout=(1,2), legend=:bottomright)
plot!(size=(950,400))
```

```julia
#Know the proportion each class
class_freq = combine(groupby(data, :class), :class=>length)
```

```julia
bar(class_freq.class, class_freq.class_length, title="Number of Class", label = "")
plot!(size=(950,400))
```

## Using Scikit-Learn

```julia
#Import ScikiLearn from Python Language in your machine
#Make sure you already installed python dan ScikitLearn pakcage in Julia
using ScikitLearn
using ScikitLearn.CrossValidation: train_test_split
#@sk_import models_selection: train_test_split
@sk_import preprocessing: StandardScaler
@sk_import tree: (DecisionTreeClassifier, export_graphviz)
@sk_import ensemble: (RandomForestClassifier, AdaBoostClassifier,
ExtraTreesClassifier, GradientBoostingClassifier)
@sk_import metrics: classification_report;
```

```julia
#Split the feature and target
X = Array(data[:,[1,2,3,4]])
y = Array(data[:,[5]]);
```

```julia
#Split the train and validation data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2,
stratify=y, random_state=42)    #use stratify to ensure the target have same proportion when splitted the data,
                                #and random state to make the split consistent each run
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test);
```

```julia
data_labels = DataFrame(y_train)
class_freq = combine(groupby(data_labels, :x1), :x1=>length)
```

## Decision Tree

```julia
clf = fit!(DecisionTreeClassifier(random_state=42), X_train, y_train)
export_graphviz(clf,out_file="tree_iris.dot") #You can change this to png using graphviz
```

```julia
pred = predict(clf, X_test)
println(classification_report(y_test, pred))
```

## Random Forrest

```julia
rf = fit!(RandomForestClassifier(n_estimators=1000, random_state=42), X_train, y_train)
pred = predict(rf, X_test)
println(classification_report(y_test, pred))
```

## AdaBoostClassifier

```julia
ABC = fit!(AdaBoostClassifier(n_estimators=500, learning_rate=1e-3, random_state=42), X_train, y_train)
pred = predict(ABC, X_test)
println(classification_report(y_test, pred))
```

## ExtraTreesClassifier

```julia
etc = fit!(ExtraTreesClassifier(n_estimators=1000, random_state=42), X_train, y_train)
pred = predict(etc, X_test)
println(classification_report(y_test, pred))
```

## GradientBoostingClassifier

```julia
gbc = fit!(GradientBoostingClassifier(n_estimators=1000, learning_rate=1e-3, random_state=42), X_train, y_train)
pred = predict(gbc, X_test)
println(classification_report(y_test, pred))
```
